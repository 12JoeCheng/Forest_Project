# Share of tropical deforestation from agricultural products in 2013
----

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('share-deforestation-agricultural-products.csv')
df
```

```python
df = df.sort_values(by = 'Share of commodity-driven deforestation', ascending = False)
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(25,10))

activity = []
for x in df['Entity']:
    activity.append(x)

percentage = []
for x in df['Share of commodity-driven deforestation']:
    percentage.append(x)

activity.reverse()
percentage.reverse()

plt.rc('ytick', labelsize=20)
plt.rc('xtick', labelsize=20)

plt.barh(activity, percentage)

plt.title('Share of tropical deforestation from agricultural products in 2013', fontsize = 30)
plt.xlabel('Percentage', fontsize = 18)

plt.tight_layout()
plt.show()
```
