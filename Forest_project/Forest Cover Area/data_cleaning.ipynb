# What share of each country is forested?
----
>### Forest Cover Area

>>#### the term **‘Forest Cover Area’**  refers to all lands more than 1 hectare in area, having a tree canopy density of more than 10%. Indicating the presence of trees over any land.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
slf_df = pd.read_csv('forest-area-as-share-of-land-area.csv')
slf_df.head()
```

#

### Countries

```python
filt1 = (slf_df['Code'].isna())
countries_slf_df = slf_df[~filt1]
countries_slf_df.head()
```

```python
filt2 = (countries_slf_df['Year'] == 2020)
ConvertedCountries2020_df = countries_slf_df[filt2].sort_values(by = 'Forest cover', ascending= False)
ConvertedCountries2020_df
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(14,8))

countries = []
percentage = []

for x in ConvertedCountries2020_df['Entity'][0:19]:
    countries.append(x)

for x in ConvertedCountries2020_df['Forest cover'][0:19]:
    percentage.append(x)


countries.reverse()
percentage.reverse()

plt.rc('ytick', labelsize=20)
plt.rc('xtick', labelsize=20)

plt.barh(countries, percentage)

plt.title('2020 Forest Cover Area in Countries', fontsize = 35)
plt.xlabel('Percentage', fontsize = 20)

plt.tight_layout()
plt.show()
```

#

### Cover Ratio Change from 2000 to 2020 in Countries

```python
countries_slf_df
```

```python
countries_gb_df = countries_slf_df.groupby(['Entity'])
```

```python
countries = list(countries_slf_df['Entity'].unique())
len(countries)
```

```python
result = []

for x in countries:
    data = countries_gb_df.get_group(f'{x}')

    filt = (data['Year'] == 2000)
    rate_2000 = data.loc[filt, 'Forest cover']

    filt = (data['Year'] == 2020)
    rate_2020 = data.loc[filt, 'Forest cover']

    try:
        rate = (float(rate_2020) - float(rate_2000)) * 100/float(rate_2000)
    except:
        rate = 0

    result.append(rate)

```

```python
len(result)
```

```python
yoy_countries_df = pd.DataFrame({'Entity': countries,
                                    'RatioChangePercentage': result})

yoy_countries_df.set_index('Entity', inplace = True)
```

```python
yoy_countries_df.sort_values(by = 'RatioChangePercentage', ascending = False, inplace = True)
```

```python
yoy_countries_df
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(14,8))

countries = []
for x in yoy_countries_df.index[0:9]:
    countries.append(x)

percentage_change = []
for x in yoy_countries_df['RatioChangePercentage'][0:9]:
    percentage_change.append(x)

countries.reverse()
percentage_change.reverse()

plt.rc('ytick', labelsize=15)
plt.rc('xtick', labelsize=15)

plt.barh(countries, percentage_change)

plt.title('From 2000 to 2020 Cover Area Ratio Most Change Percentage', fontsize = 30)
plt.xlabel('Percentage', fontsize = 18)

plt.tight_layout()
plt.show()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(14,8))

countries = []
for x in yoy_countries_df.index[:-11:-1]:
    countries.append(x)

percentage_change = []
for x in yoy_countries_df['RatioChangePercentage'][:-11:-1]:
    percentage_change.append(x)

countries.reverse()
percentage_change.reverse()

plt.rc('ytick', labelsize=15)
plt.rc('xtick', labelsize=15)

plt.barh(countries, percentage_change)

plt.title('From 2000 to 2020 Cover Area Ratio Least Change Percentage', fontsize = 30)
plt.xlabel('Percentage', fontsize = 18)

plt.tight_layout()
plt.show()
```

### Cover Ratio Change from 2000 to 2020 in Continents

```python
temp_1 = ['Asia', 'Northern America', 'South America', 'Europe', 'Oceania', 'Africa']
temp_2 = ['Asia', 'Americas', 'Europe', 'Africa']

ConvertedContinents_slf_df_3 = ConvertedContinents_slf_df[ConvertedContinents_slf_df['Entity'].isin(temp_1)]
ConvertedContinents_slf_df_4 = ConvertedContinents_slf_df[~ConvertedContinents_slf_df['Entity'].isin(temp_2)]

continents_3 = list(ConvertedContinents_slf_df_3['Entity'].unique())
continents_4 = list(ConvertedContinents_slf_df_4['Entity'].unique())
```

```python
continents_gb_df_3 = ConvertedContinents_slf_df_3.groupby(['Entity'])
continents_gb_df_4 = ConvertedContinents_slf_df_4.groupby(['Entity'])
```

```python
result = []

for x in continents_3:
    data = continents_gb_df_3.get_group(f'{x}')

    filt = (data['Year'] == 2000)
    rate_2000 = data.loc[filt, 'Forest cover']

    filt = (data['Year'] == 2020)
    rate_2020 = data.loc[filt, 'Forest cover']

    try:
        rate = (float(rate_2020) - float(rate_2000)) * 100/float(rate_2000)
    except:
        rate = 0

    result.append(rate)

len(result)
```

```python
yoy_continents_df_3 = pd.DataFrame({'Entity': continents_3,
                                    'RatioChangePercentage': result})

yoy_continents_df_3.set_index('Entity', inplace = True)

yoy_continents_df_3.sort_values(by = 'RatioChangePercentage', ascending = False, inplace = True)

yoy_continents_df_3
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(14,8))

continents = []
for x in yoy_continents_df_3.index:
    continents.append(x)

percentage_change = []
for x in yoy_continents_df_3['RatioChangePercentage']:
    percentage_change.append(x)

continents.reverse()
percentage_change.reverse()

plt.rc('ytick', labelsize=15)
plt.rc('xtick', labelsize=15)

plt.barh(continents, percentage_change)

plt.title('From 2000 to 2020 each Continents Cover Area Ratio Change Percentage', fontsize = 25)
plt.xlabel('Percentage', fontsize = 18)

plt.tight_layout()
plt.show()
```
